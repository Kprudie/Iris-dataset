```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = sns.load_dataset("iris")  # Built-in dataset

# Preview data
print("First 5 rows:")
print(df.head())

# Check for missing values
print("\nMissing values:")
print(df.isnull().sum())

# Clean the data (this ensures df_cleaned is always defined)
if df.isnull().sum().sum() > 0:
    df_cleaned = df.dropna()
else:
    df_cleaned = df.copy()

# Basic statistics
print("\nDescriptive Statistics:")
print(df_cleaned.describe())

# Group by species
grouped_means = df_cleaned.groupby("species").mean()
print("\nGroup-wise Mean:")
print(grouped_means)

# Line chart (index vs petal_length)
plt.figure(figsize=(8,4))
df_cleaned['petal_length'].plot(kind='line', title='Petal Length Over Index')
plt.xlabel('Index')
plt.ylabel('Petal Length')
plt.grid(True)
plt.show()

# Bar chart: average petal length per species
plt.figure(figsize=(6,4))
grouped_means['petal_length'].plot(kind='bar', title='Avg Petal Length per Species')
plt.ylabel('Petal Length')
plt.grid(True)
plt.show()

# Histogram: sepal length
plt.figure(figsize=(6,4))
plt.hist(df_cleaned['sepal_length'], bins=10, color='skyblue', edgecolor='black')
plt.title('Sepal Length Distribution')
plt.xlabel('Sepal Length')
plt.ylabel('Count')
plt.grid(True)
plt.show()

# Scatter plot: sepal_length vs petal_length
plt.figure(figsize=(6,4))
sns.scatterplot(data=df_cleaned, x='sepal_length', y='petal_length', hue='species')
plt.title('Sepal vs Petal Length')
plt.grid(True)
plt.show()
```
